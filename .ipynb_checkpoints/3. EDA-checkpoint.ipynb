```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
import plotly.express as px
pd.options.display.float_format = '{:.2f}'.format
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
males = pd.read_csv('1. Data/Data/male_names.csv', parse_dates =['year'], index_col="year")
females = pd.read_csv('1. Data/Data/female_names.csv', parse_dates =['year'], index_col="year")
```

```python
males.info()
```

```python
females.info()
```

### Plot Total Names per Year Per Sex

```python
plt.figure(figsize=(15,10))
plt.plot(males.groupby(males.index)["count"].sum(), label="total number of male names")
plt.plot(females.groupby(females.index)["count"].sum(), label="total number of female names")
plt.ylabel("Count", fontsize=16)
plt.xlabel("Year", fontsize=16)
plt.legend()
plt.title("Total Number of Names By Sex Yearly", fontsize=18);
```

### Total Number of Unique Names per Year per Sex

```python
plt.figure(figsize=(15,10))
plt.plot(males.groupby(males.index)["name"].nunique(), label="total number of unique male names")
plt.plot(females.groupby(females.index)["name"].nunique(), label="total number of unique female names")
plt.legend()
plt.ylabel("Number of Names", fontsize=16)
plt.xlabel("Year", fontsize=16)
plt.title("Total Unique Names per Sex Yearly", fontsize=18);
```

### Total number of unique names per sex in the data

```python
len(list(males["name"].unique())), len(list(females["name"].unique()))
```

```python
plt.figure(figsize=(10,10))
plt.title('Number of Unique Names per Sex', fontdict={'fontsize':20})
plt.ylabel('Number of Unique Names', fontsize=16)
plt.xlabel('Sex', fontsize=16);
plt.bar(x=['Male Names', 'Female Names'], height=[len(list(males["name"].unique())), len(list(females["name"].unique()))], color=['blue','crimson']);
```

### based on EDA--initially started with names of at least 100 name mentions from 1910 to 2020 to start

```python
males_100_1910 = pd.read_csv('1. Data/Data/males100_1910.csv', parse_dates =['year'], index_col="year")
females_100_1910 = pd.read_csv('1. Data/Data/females100_1910.csv', parse_dates =['year'], index_col="year")
```

```python
len(list(males_100_1910["name"].unique())), len(list(females_100_1910["name"].unique()))
```

```python
plt.figure(figsize=(15,10))
plt.plot(males_100_1910.groupby(males_100_1910.index)["name"].nunique(), label="total number of unique male names with at least 100 mentions")
plt.plot(females_100_1910.groupby(females_100_1910.index)["name"].nunique(), label="total number of unique female names with at least 100 mentions")
plt.legend()
plt.title("Total Unique Names with at least 100 mentions per Sex Yearly");
```

```python
plt.figure(figsize=(15,10))
plt.plot(males_100_1910.groupby(males_100_1910.index)["count"].sum(), label="total number of male names with at least 100 mentions")
plt.plot(females_100_1910.groupby(females_100_1910.index)["count"].sum(), label="total number of female names with at least 100 mentions")
plt.legend()
plt.title("Total Number of Names with at least 100 mentions By Sex Yearly");
```

```python
popular_males=pd.read_csv('1. Data/Data/popular_males.csv', parse_dates =['year'], index_col="year")
popular_females=pd.read_csv('1. Data/Data/popular_females.csv', parse_dates =['year'], index_col="year")
```

```python
plt.figure(figsize=(15,10))
plt.plot(popular_males.groupby(popular_males.index)["name"].nunique(), label="total number of popular individual male names")
plt.plot(popular_females.groupby(popular_females.index)["name"].nunique(), label="total number of popular individual female names")
plt.legend()
plt.ylabel("Number of Names", fontsize=16)
plt.xlabel("Year", fontsize=16)
plt.title("Total Number of Individual 'Popular' Names", fontsize=18);
```

```python
plt.figure(figsize=(15,10))
plt.plot(popular_males.groupby(popular_males.index)["count"].sum(), label="total count for the most popular male names")
plt.plot(popular_females.groupby(popular_females.index)["count"].sum(), label="total count for the most popular female names")
plt.legend()
plt.ylabel("Count", fontsize=16)
plt.xlabel("Year", fontsize=16)
plt.title("Total Count for 'Popular' Names by Year" , fontsize=18);
```

```python

names = ["Theodore", "Marshall", "Barney"]
mask_names=males["name"].isin(names)
df=males[mask_names]["1920":"2020"]


fig = px.line(df, x=df.index, y="count", color="name", title = "'Trends for the Male Names in 'How I Met Your Mother'")
fig.add_vrect(x0="2005", x1="2014", annotation_text="HIMYM", annotation_position="top left",
              annotation=dict(font_size=20, font_family="Times New Roman"),
              fillcolor="blue", opacity=0.25, line_width=0)
fig.update_layout(width = 700, height=600, xaxis = dict(title = 'Year', showticklabels=True), yaxis=dict(title="Count" , showticklabels=True))
```
