```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
import plotly.express as px
```

```python
males = pd.read_csv('./Data/male_names.csv', parse_dates =['year'], index_col="year")
females = pd.read_csv('./Data/female_names.csv', parse_dates =['year'], index_col="year")
```

```python
males.info()
```

```python
females.info()
```

### Plot Total Names per Year Per Sex

```python
plt.figure(figsize=(15,10))
plt.plot(males.groupby(males.index)["count"].sum(), label="total number of male names")
plt.plot(females.groupby(females.index)["count"].sum(), label="total number of female names")
plt.legend()
plt.title("Total Number of Names By Sex Yearly");
```

### Total Number of Unique Names per Year per Sex

```python
plt.figure(figsize=(15,10))
plt.plot(males.groupby(males.index)["name"].nunique(), label="total number of unique male names")
plt.plot(females.groupby(females.index)["name"].nunique(), label="total number of unique female names")
plt.legend()
plt.title("Total Unique Names per Sex Yearly");
```

### Total number of unique names per sex in the data

```python
len(list(males["name"].unique())), len(list(females["name"].unique()))
```

```python
plt.figure(figsize=(10,10))
plt.title('Number of Unique Names per Sex', fontdict={'fontsize':20})
plt.ylabel('Number of Unique Names', fontsize=16)
plt.xlabel('Sex', fontsize=16);
plt.bar(x=['Male Names', 'Female Names'], height=[len(list(males["name"].unique())), len(list(females["name"].unique()))], color=['blue','crimson']);
```

### based on EDA--initially started with names of at least 100 name mentions from 1910 to 2020 to start

```python
males_100_1910 = pd.read_csv('./Data/males100_1910.csv', parse_dates =['year'], index_col="year")
females_100_1910 = pd.read_csv('./Data/females100_1910.csv', parse_dates =['year'], index_col="year")
```

```python
len(list(males_100_1910["name"].unique())), len(list(females_100_1910["name"].unique()))
```
